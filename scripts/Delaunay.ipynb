We load the neccessary libraries

```python
from scipy.spatial import Delaunay
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
import random
import numpy as np
from google.colab import files
```

We generate a random sample of points and call the Delaunay function

```python
SIZE = 800
points = np.array([(random.randrange(SIZE), random.randrange(SIZE))
               for i in range(80)])
tri = Delaunay(points)
```

We plot the results

```python
plt.triplot(points[:,0], points[:,1], tri.simplices)
plt.plot(points[:,0], points[:,1], 'o')
#plt.savefig("Delaunay.pdf")
#files.download("Delaunay.pdf")
plt.show()
```

We can do the exact same for the corresponding dual Voronoi diagram

```python
vor = Voronoi(points)
fig = voronoi_plot_2d(vor)
#plt.savefig("Voronoi.pdf")
#files.download("Voronoi.pdf")
plt.show()
```
